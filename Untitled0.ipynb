```python
from pytrends.request import TrendReq
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import pandas as pd
```

```python
# Membuat sesi Spark
spark = SparkSession.builder.master("local[*]").getOrCreate()

# Mengambil data Google Trends dengan pytrends
pytrends = TrendReq(hl='en-US', tz=360)

# Kata kunci tren terkait sepeda
kw_list = ["Sepeda Listrik", "Sepeda Gunung", "Sepeda Balap", "Sepeda Lipat", "Sepeda Anak"]

# Mengambil data tren untuk 5 tahun terakhir
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')

# Mendapatkan interest over time (minat seiring waktu)
data = pytrends.interest_over_time()

# Menghapus kolom 'isPartial'
data = data.drop(columns=['isPartial'])

# Mengonversi data ke DataFrame Spark
df_spark = spark.createDataFrame(data.reset_index())
df_spark.show(5)  # Menampilkan 5 baris pertama
```

```python
df_spark.write.csv("google_trends_data_sepeda.csv")
```

```python
import matplotlib.pyplot as plt

# Mengubah DataFrame PySpark menjadi Pandas untuk visualisasi
df_pandas = df_spark.toPandas()

# Visualisasi tren dari data
df_pandas.plot(x='date', y=kw_list, figsize=(10,6))
plt.title('Tren Sepeda di Google Trends')
plt.xlabel('Tanggal')
plt.ylabel('Minat Pencarian')
plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
```

```python
# Membuat sesi Spark
spark = SparkSession.builder.master("local[*]").getOrCreate()

# Contoh data penjualan (Harga dan Penjualan)
data = [(1000, 10), (1500, 15), (1200, 12), (3000, 20), (4000, 25), (3500, 22)]
columns = ["Harga", "Penjualan"]

# Membuat DataFrame
df = spark.createDataFrame(data, columns)

# Menggabungkan kolom Harga dan Penjualan menjadi fitur untuk K-Means
assembler = VectorAssembler(inputCols=["Harga", "Penjualan"], outputCol="features")
df_features = assembler.transform(df)

df_features.show()  # Melihat data hasil transformasi
```

```python
# Membuat model K-Means dengan 3 klaster
kmeans = KMeans(k=3, seed=1, featuresCol="features", predictionCol="prediction")
model = kmeans.fit(df_features)

# Menerapkan model untuk memprediksi klaster
predictions = model.transform(df_features)

# Menampilkan hasil klaster
predictions.select("Harga", "Penjualan", "prediction").show()
```

```python
# Evaluasi clustering menggunakan Silhouette Score
evaluator = ClusteringEvaluator(featuresCol="features", predictionCol="prediction")

silhouette = evaluator.evaluate(predictions)
print(f"Silhouette Score: {silhouette}")
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

# Membuat sesi Spark
spark = SparkSession.builder.master("local[*]").getOrCreate()

# Contoh data penjualan
data = [(1000, 10), (1500, 15), (1200, 12), (3000, 20), (4000, 25), (3500, 22)]
columns = ["Harga", "Penjualan"]

# Membuat DataFrame
df = spark.createDataFrame(data, columns)

# Menggabungkan kolom Harga dan Penjualan menjadi fitur untuk K-Means
assembler = VectorAssembler(inputCols=["Harga", "Penjualan"], outputCol="features")
df_features = assembler.transform(df)

# Membuat model K-Means dengan 3 klaster
kmeans = KMeans(k=3, seed=1, featuresCol="features", predictionCol="prediction")
model = kmeans.fit(df_features)

# Menerapkan model untuk memprediksi klaster
predictions = model.transform(df_features)

# Menampilkan hasil klaster
predictions.select("Harga", "Penjualan", "prediction").show()

# Evaluasi clustering menggunakan Silhouette Score
evaluator = ClusteringEvaluator(featuresCol="features", predictionCol="prediction")

silhouette = evaluator.evaluate(predictions)
print(f"Silhouette Score: {silhouette}")
```
